```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line Plot

```python
# Line Plot

x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1,color="red", label="Apple", marker='o')
plt.plot(x,y2,color="green", label="Kiwi", linestyle='dashed', marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend()
plt.show()
```

```python
prices = np.array([1,2,3,4])**2
print(prices)
```

```python
plt.plot(prices)
plt.show()
```

### Scatter Plots

```python
plt.figure(figsize=(5,5))

plt.scatter(x,y1,color="red", label="Apple", marker='^')
plt.scatter(x,y2,color="green", label="Kiwi", linestyle='dashed')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot : Prices over time")
plt.legend()
plt.show()
```

### Bar Graphs

```python
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates-0.25,[10,20,15], width=0.5, label = 'Current Year', tick_label = ["Gold", "Platinum", "Silver"]) # Current Year
plt.bar(x_coordinates+0.25,[20,10,12], width=0.5, label = 'Next Year') # Next Year
plt.xlabel("Metal")
plt.ylabel("Price")
plt.title("Metal Price Comparision")

plt.ylim(0,25)
plt.xlim(-2,5)

plt.legend()
plt.show()
```

### Pie Charts

```python
subjects = "Maths", "Chem", "Phy", "Eng"
weightage = [20, 10, 15, 5]

plt.pie(weightage, labels=subjects)
plt.show()
```

```python
#Cutting out the slice
plt.pie(weightage, labels=subjects, explode = (0,0.1,0,0))
plt.show()
```

```python
# Percentage
plt.pie(weightage, labels=subjects, explode = (0,0.1,0,0), autopct = '%1.1f%%', shadow = True)
plt.show()
```

### Histogram

```python
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)
X2 = np.round(Xsn*15 + 40)

print(X,X2)
```

```python
plt.hist(X, alpha=0.9, label = "Phy")
plt.hist(X2, alpha=0.8, label = "Maths")
plt.xlabel("Marks Range")
plt.ylabel("Prob/ Freq Count of Students")
plt.title("Histogram")
plt.legend()
plt.show()
```
