# Data Visualisation Challenge

Given a Movies Dataset, your task is to make a visualization that :

- Plot the Length of Movies Title Names on X-axis
- Frequency Counts on Y-axis ie Number of movies having 'x' characteristics in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read a CSV File
df = pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
df.columns # Displays all  the column present
```

```python
titles = list(df.get('movie_title')) #Extracts movie_name from csv and convert into elements of list
```

```python
print(titles[:5])
```

```python
print(titles[0])
```

```python
# Converting length of movie title as keys and no of movies as values
freq_titles={}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    
    else:
        freq_titles[length]+=1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))

#print(X,Y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualisation Problem")
plt.show()
```
